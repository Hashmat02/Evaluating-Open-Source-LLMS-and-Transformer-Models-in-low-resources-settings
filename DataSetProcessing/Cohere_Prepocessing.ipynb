```python
import json
from google.cloud import translate_v2 as translate

# Initialize Google Cloud Translation API client
client = translate.Client.from_service_account_json('sunlit-descent-445008-a6-ebadbc918b1c.json')

# File name in the same directory
file_name = 'aya_eng.jsonl'

# Categories to filter
categories = [
    "Discrimination & Injustice",
    "Bullying & Harassment",
    "Harms of Representation Allocation and Quality of Service"
]

# Target languages for translation
languages = {
    'ur': 'Urdu',
    'sd': 'Sindhi',  # Sindhi language
    'ps': 'Pashto'
}

# Dictionary to store results
category_prompts = {category: [] for category in categories}

# Step 1: Read and Extract Prompts
with open(file_name, 'r', encoding='utf-8') as file:
    for line in file:
        json_obj = json.loads(line)
        harm_categories = json_obj.get('harm_category', [])
        for category in categories:
            if category in harm_categories and len(category_prompts[category]) < 10:
                category_prompts[category].append(json_obj['prompt'])

# Step 2: Translate the Prompts
translated_prompts = {
    category: {'original': [], **{lang: [] for lang in languages}}
    for category in categories
}

for category, prompts in category_prompts.items():
    for prompt in prompts:
        translated_prompts[category]['original'].append(prompt)
        for lang_code, lang_name in languages.items():
            # Translate prompt using Google API
            result = client.translate(prompt, target_language=lang_code)
            translated_prompts[category][lang_code].append(result['translatedText'])

# Step 3: Print Original and Translated Prompts
for category, translations in translated_prompts.items():
    print(f"\nCategory: {category}")
    print("=" * 60)
    for i, original_prompt in enumerate(translations['original'], start=1):
        print(f"\nPrompt {i}:")
        print(f"Original (English): {original_prompt}")
        for lang_code, lang_name in languages.items():
            print(f"{lang_name} Translation: {translations[lang_code][i-1]}")
```
